# Poisson Distribution

## Derivation of Homogeneous Poisson Distribution

### Bottom up Derivation

Taken from https://www.pp.rhul.ac.uk/~cowan/stat/notes/PoissonNote.pdf

We want to derive the probability that $n$ events occur in a time $t$, $p(n;t)$, under the assumption that *each event is independent*. First, let's consider small time intervals, $\delta t$, such that at most 1 event can occur within $\delta t$. We suppose that the probability that 1 event occurs in $\delta t$ is given by:

\begin{equation}
    p(1;\delta t) = \lambda \delta t \;\;\; , \;\;\; p(0;\delta t) = 1 - \lambda \delta t
\end{equation}

where $\lambda$ *is constant* (hence homogeneous). To start, the probability that no events occur in a time $t + \delta t$ is given by

\begin{align}
    p(0;t+\delta t) &= p(0;t)p(0;\delta t) \;\;\;\;\;\;\; (\mathrm{independent \; probabilities}) \\
    &= p(0;t) (1 - \lambda \delta t) \\
    &= p(0;t) - \lambda p(0;t) \delta t. \\
\end{align}

Rearranging, we get

\begin{align}
    \frac{p(0;t+\delta t) - p(0;t)}{\delta t} &= -\lambda p(0;t) \\
    \frac{dp(0;t)}{dt} &= -\lambda p(0;t) \\
    p(0;t) &= Ce^{-\lambda t} = e^{-\lambda t},
\end{align}

where $C = 1$ since $p(0;0) = 1$. Now we can play the same game, but for $n$ events. The probability of $n$ events occuring in a time interval $t + \delta t$ is given by:

\begin{align}
    p(n;t+\delta t) &= p(n;t)p(0;\delta t) + p(n-1;t)p(1;\delta t).
\end{align}

Terms such as $p(n > 1;\delta t) = 0$, since we made the assumption that only 1 event could occur in a time $\delta t$. This expression becomes:

\begin{align}
    p(n;t+\delta t) &= p(n;t)(1 - \lambda \delta t) + \lambda p(n-1;t) \delta t. \\
    \frac{p(n;t+\delta t) - p(n;t)}{\delta t} &= -\lambda p(n;t) + \lambda p(n-1;t) \\
    \frac{dp(n;t)}{dt} + \lambda p(n;t) &= \lambda p(n-1; t).
\end{align}

This expression is in the form $y'(t) + P(t)y(t) = Q(t)$, which is solved by an integrating factor $M(t)$, s.t. $(M(t)y(t))' = M(t)Q(t)$. Here, $M'(t) = M(t)P(t)$, leading to $M(t) = D\,\mathrm{exp}(\int dt P(t))$. In our case, $P(t) = Q(t) = \lambda$, so $M(t) = D e^{\lambda t}$. Using, this we can re-write our differential equation as:

\begin{align}
    \frac{d}{dt}\left( e^{\lambda t} p(n;t) \right) &= \lambda e^{\lambda t}p(n-1;t)
\end{align}

Let's now look at the $n = 1$ case:

\begin{align}
    \frac{d}{dt}\left( e^{\lambda t} p(1;t) \right) &= \lambda e^{\lambda t}p(0;t). \\
    \frac{d}{dt}\left( e^{\lambda t} p(1;t) \right) &= \lambda e^{\lambda t}e^{-\lambda t} = \lambda.
\end{align}

Solving for $p(1;t)$, we get:

\begin{align}
    e^{\lambda t} p(1;t) &= \int dt \lambda. \\
    p(1;t) &= \lambda t e^{-\lambda t} + ce^{-\lambda t} = \lambda t e^{-\lambda t}
\end{align}

where $c = 1$ since $p(1;0) = 0$. Similarly, for $n = 2$, we get:

\begin{align}
    p(2;t) &= \frac{(\lambda t)^2}{2} e^{-\lambda t}.
\end{align}

We now assert that the probability of $n$ events occuring in a time $t$ is given by:

\begin{align}
    p(n;t) &= \frac{(\lambda t)^n}{n!} e^{-\lambda t}
\end{align}

which we will prove by induction. For $n+1$ events, we have:

\begin{align}
    \frac{d}{dt}\left( e^{\lambda t} p(n+1;t) \right) &= \lambda e^{\lambda t}p(n;t). \\
    \frac{d}{dt}\left( e^{\lambda t} p(n+1;t) \right) &= \frac{\lambda^{n+1}}{n!}t^n . \\
    p(n+1;t) &= \frac{(\lambda t)^{n+1}}{(n+1)!}e^{-\lambda t}
\end{align}

completing the proof.

The expectation number of events in a time $t$ can be computed as:

\begin{align}
    <n> &= \sum_{n=0}^\infty n p(n;t) \\
    &= \sum_{n=0}^\infty n \frac{\nu^n}{n!}e^{-\nu} \\
    &= \sum_{n=1}^\infty n \frac{\nu^n}{n!}e^{-\nu} \\
    &= e^{-\nu} \sum_{n=1}^\infty \frac{\nu^n}{(n-1)!} \\
    &= e^{-\nu} \nu \sum_{m=0}^\infty \frac{\nu^m}{(m)!} \;\;\;\;\;\; (m = n - 1) \\
    &= \nu e^{-\nu} e^{\nu} \\
    &= \nu
\end{align}

where $\nu = \lambda t$. Likewise, the standard deviation can be computed as $\sigma^2 = <n^2> - <n>^2$:

\begin{align}
    <n^2> &= \sum_{n=0}^\infty n^2 p(n;t) \\
    &= \sum_{n=1}^\infty n^2 \frac{\nu^n}{n!}e^{-\nu} \\
    &= e^{-\nu} \sum_{n=1}^\infty n \frac{\nu^n}{(n-1)!} - \left( e^{-\nu} \sum_{n=1}^\infty \frac{\nu^n}{(n-1)!} - e^{-\nu} \sum_{n=1}^\infty \frac{\nu^n}{(n-1)!} \right)\\
    &= e^{-\nu} \sum_{n=2}^\infty (n-1) \frac{\nu^n}{(n-1)!} + e^{-\nu} \sum_{n=1}^\infty \frac{\nu^n}{(n-1)!} \\
    &= e^{-\nu} \nu^2 \sum_{n=2}^\infty \frac{\nu^{n-2}}{(n-2)!} + e^{-\nu} nu \sum_{n=1}^\infty \frac{\nu^{n-1}}{(n-1)!} \\
    &= e^{-\nu} \nu^2 \sum_{k=0}^\infty \frac{\nu^{k}}{(k)!} + e^{-\nu} nu \sum_{m=0}^\infty \frac{\nu^{m}}{(m)!} \;\;\;\;\;\; (k = n - 2) \\
    &= \nu^2 e^{-\nu} e^{\nu} + \nu e^{-\nu} e^{\nu} \\
    &= \nu^2 + \nu
\end{align}

so $\sigma = \sqrt{\nu^2 + \nu - \nu^2} = \sqrt{\nu}$.


### Poisson Distribution as a limit of Binomial Distribution

Very naturally, we can arrive at the Poisson distribution directly from the Binomial distribution. The Binomial distribution says that for $n$ events, each of which have a probability $p$ of success, then the probability that $k$ of these events are succesful is given by:

\begin{align}
    B(k;n,p) = \frac{n!}{(n-k)!\,k!}\,p^k(1-p)^{n-k}.
\end{align}

Now to the Poisson distibution can be obtained, by considering the limit where the number of events (i.e. the number of time intervals $\delta t$) $n \to \infty$ and while keeping the rate $\lambda = np$ fixed. Now, we can eliminate $p$ from $B(k;n,p)$ and only consider the large $n$ limit. In doing so, we propose:

\begin{align}
    p(\lambda, k) = \lim_{n \to \infty} B(k;n,\lambda) = \lim_{n \to \infty} \frac{n!}{(n-k)!\,k!}\,\frac{\lambda^k}{n^k}(1-\frac{\lambda}{n})^{n-k}.
\end{align}

Taking the limit of various pieces:

\begin{align}
    \lim_{n \to \infty} \frac{n!}{(n-k)!}\,\frac{1}{n^k} &= \lim_{n \to \infty} \frac{n(n-1)\cdots{n-k+1}}{n^k} = \lim_{n \to \infty} \frac{n}{n}\frac{n-1}{n}\cdots\frac{n-k+1}{n} = 1 \\
    \lim_{n \to \infty} (1-\frac{\lambda}{n})^{-k} = 1 \\
    \lim_{n \to \infty} (1-\frac{\lambda}{n})^{n} = e^{-\lambda} \\
\end{align}

Plugging these in, we reduce to 

\begin{align}
    p(\lambda, k) = \lim_{n \to \infty} B(k;n,\lambda) = \frac{\lambda^k}{k!}e^{-\lambda}
\end{align}

which gives us the Poisson distribution as expected.

## Inhomogeneous Poisson Distribution

The derivation is identical to that of the homogeneous Poisson distribution, except that integrals over $\lambda$ must be kept in place. One finds that in the original formula, it suffices to replace $\lambda t \to \int_0^t \lambda(\tau) d\tau$:

\begin{align}
    p(n;t) = \frac{\bigg[ \int_0^t \lambda(\tau) d\tau \bigg]^n}{n!} e^{-\int_0^t \lambda(\tau) d\tau}
\end{align}

## Plots

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

```python
plt.rcParams.update({
        'axes.grid': False,
        'axes.labelsize': 12,
        'axes.linewidth': 0.8,
        'axes.titlesize': 12,
        'font.size': 10,
        'legend.fontsize': 12,
        'xtick.labelsize': 12,
        'ytick.labelsize': 12,
        'font.family': 'serif',
        'font.sans-serif': ['Bitstream Vera Sans'],
        'font.serif': ['Times New Roman'],
        'text.latex.preamble': r'\usepackage{amsmath} \usepackage{amssymb} \usepackage{amsfonts}',
        'text.usetex': True,
        'patch.force_edgecolor': True,
        'figure.dpi': 100,
        'savefig.dpi': 100,
    }
)
```

```python
n = np.linspace(0, 20, 21)  # Number of events (integers).
nu = np.linspace(1, 10, 10)   # Expected number of events \nu = \lambda * t, where lambda is the rate and t is the time interval.

for i in nu:
    p = ( (i**n) / factorial(n) ) * np.exp(-i)
    plt.plot(n, p, label=r'$\nu =$ '+ f'{i:.1f}')

plt.xlabel(r'$n$')
plt.ylabel(r'$p(n;\nu)$')
plt.legend()
plt.ylim(0, 0.5)
plt.xlim(0, 20)
plt.show()
```
